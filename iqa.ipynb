```python
# Normalize 'Operator' column: strip spaces and use title case
df['Operator'] = df['Operator'].str.strip().str.title()

# If you want to map specific variants:
# df['Operator'] = df['Operator'].replace({'MUGERWA': 'Mugerwa', 'SIMON': 'Simon'})

# Check the result
print(df['Operator'].value_counts(dropna=False))
```

```python
# Get audio files annotated by Simon
simon_audios = set(df[df['Operator'] == 'Simon']['Audio File'])

# Get audio files annotated by Chris
chris_audios = set(df[df['Operator'] == 'Chris']['Audio File'])

# Find shared audio files
shared_audios = simon_audios & chris_audios

print("Audio files annotated by both Simon and Chris:")
print(shared_audios)
```

```python
print("Number of shared audio files:", len(shared_audios))
```

```python
# For each shared audio, show label counts and label types per operator
for audio in shared_audios:
    print(f"\nAudio File: {audio}")
    for operator in ['Simon', 'Chris']:
        op_df = df[(df['Audio File'] == audio) & (df['Operator'] == operator)]
        label_counts = op_df['Label'].value_counts()
        print(f"  Operator: {operator}")
        print(f"    Number of labels: {len(op_df)}")
        print(f"    Labels used: {list(label_counts.index)}")
        print(f"    Label counts:\n{label_counts.to_string()}")
```

```python
from collections import defaultdict

for audio in shared_audios:
    print(f"\nAudio File: {audio}")
    simon_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Simon')]
    chris_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Chris')]
    all_labels = set(simon_df['Label']).union(set(chris_df['Label']))
    for label in all_labels:
        simon_spans = set(zip(
            simon_df[simon_df['Label'] == label]['Start'],
            simon_df[simon_df['Label'] == label]['Stop']
        ))
        chris_spans = set(zip(
            chris_df[chris_df['Label'] == label]['Start'],
            chris_df[chris_df['Label'] == label]['Stop']
        ))
        matches = simon_spans & chris_spans
        simon_only = simon_spans - chris_spans
        chris_only = chris_spans - simon_spans
        print(f"  Label: {label}")
        print(f"    Exact matches: {len(matches)}")
        print(f"    Simon only: {len(simon_only)}")
        print(f"    Chris only: {len(chris_only)}")
        if simon_only:
            print(f"      Simon only spans: {simon_only}")
        if chris_only:
            print(f"      Chris only spans: {chris_only}")
```

```python
summary = []
for audio in shared_audios:
    simon_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Simon')]
    chris_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Chris')]
    summary.append({
        "Audio File": audio,
        "Simon Count": len(simon_df),
        "Chris Count": len(chris_df),
        "Simon Labels": set(simon_df['Label']),
        "Chris Labels": set(chris_df['Label'])
    })
summary_df = pd.DataFrame(summary)
display(summary_df)
```

```python
match_report = []
for audio in shared_audios:
    simon_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Simon')]
    chris_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Chris')]
    all_labels = set(simon_df['Label']).union(set(chris_df['Label']))
    for label in all_labels:
        simon_spans = set(zip(
            simon_df[simon_df['Label'] == label]['Start'],
            simon_df[simon_df['Label'] == label]['Stop']
        ))
        chris_spans = set(zip(
            chris_df[chris_df['Label'] == label]['Start'],
            chris_df[chris_df['Label'] == label]['Stop']
        ))
        matches = simon_spans & chris_spans
        simon_only = simon_spans - chris_spans
        chris_only = chris_spans - simon_spans
        match_report.append({
            "Audio File": audio,
            "Label": label,
            "Exact Matches": len(matches),
            "Simon Only": len(simon_only),
            "Chris Only": len(chris_only),
            "Simon Only Spans": list(simon_only),
            "Chris Only Spans": list(chris_only)
        })
match_df = pd.DataFrame(match_report)
display(match_df)
```

```python
# Define a threshold for "near" (e.g., 1 second)
threshold = 0.06

near_match_report = []
for audio in shared_audios:
    simon_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Simon')]
    chris_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Chris')]
    for label in set(simon_df['Label']).intersection(set(chris_df['Label'])):
        simon_spans = simon_df[simon_df['Label'] == label][['Start', 'Stop']].values
        chris_spans = chris_df[chris_df['Label'] == label][['Start', 'Stop']].values
        for s_start, s_stop in simon_spans:
            for c_start, c_stop in chris_spans:
                if abs(s_start - c_start) <= threshold and abs(s_stop - c_stop) <= threshold:
                    near_match_report.append({
                        "Audio File": audio,
                        "Label": label,
                        "Simon Start": s_start,
                        "Simon Stop": s_stop,
                        "Chris Start": c_start,
                        "Chris Stop": c_stop,
                        "Start Diff": abs(s_start - c_start),
                        "Stop Diff": abs(s_stop - c_stop)
                    })
near_match_df = pd.DataFrame(near_match_report)
display(near_match_df)
```
