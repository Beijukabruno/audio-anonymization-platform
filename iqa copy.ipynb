```python
# Define a threshold for "near" (e.g., 1 second)
threshold = 0.06

near_match_report = []
for audio in shared_audios:
    simon_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Simon')]
    chris_df = df[(df['Audio File'] == audio) & (df['Operator'] == 'Chris')]
    for label in set(simon_df['Label']).intersection(set(chris_df['Label'])):
        simon_spans = simon_df[simon_df['Label'] == label][['Start', 'Stop']].values
        chris_spans = chris_df[chris_df['Label'] == label][['Start', 'Stop']].values
        for s_start, s_stop in simon_spans:
            for c_start, c_stop in chris_spans:
                if abs(s_start - c_start) <= threshold and abs(s_stop - c_stop) <= threshold:
                    near_match_report.append({
                        "Audio File": audio,
                        "Label": label,
                        "Simon Start": s_start,
                        "Simon Stop": s_stop,
                        "Chris Start": c_start,
                        "Chris Stop": c_stop,
                        "Start Diff": abs(s_start - c_start),
                        "Stop Diff": abs(s_stop - c_stop)
                    })
near_match_df = pd.DataFrame(near_match_report)
display(near_match_df)
```

```python
df = pd.read_csv('iqa-csv/ioa_annotations_20260324_221133.csv')
df.head()

# Normalize 'Operator' column: strip spaces and use title case
df['Operator'] = df['Operator'].str.strip().str.title()

# If you want to map specific variants:
# df['Operator'] = df['Operator'].replace({'MUGERWA': 'Mugerwa', 'SIMON': 'Simon'})

# Check the result
print(df['Operator'].value_counts(dropna=False))
```

```python
def compute_overlap(a_start, a_stop, b_start, b_stop):
    return max(0, min(a_stop, b_stop) - max(a_start, b_start))        
```

```python
def match_annotations(df_simon, df_chris, overlap_threshold=0.5):
    matches = []
    matched_chris_indices = set()
    matched_simon_indices = set()

    for i, s in df_simon.iterrows():
        for j, c in df_chris.iterrows():

            if j in matched_chris_indices:
                continue  # already matched

            overlap = max(0, min(s['Stop'], c['Stop']) - max(s['Start'], c['Start']))

            if overlap >= overlap_threshold:
                matches.append({
                    's_idx': i,
                    'c_idx': j,
                    's_label': s['Label'],
                    'c_label': c['Label'],
                    's_start': s['Start'],
                    'c_start': c['Start'],
                    's_stop': s['Stop'],
                    'c_stop': c['Stop'],
                    'start_diff': abs(s['Start'] - c['Start']),
                    'stop_diff': abs(s['Stop'] - c['Stop']),
                    'label_match': s['Label'] == c['Label']
                })

                matched_simon_indices.add(i)
                matched_chris_indices.add(j)
                break

    # ✅ Now compute unmatched safely
    unmatched_simon = df_simon.loc[~df_simon.index.isin(matched_simon_indices)]
    unmatched_chris = df_chris.loc[~df_chris.index.isin(matched_chris_indices)]

    return matches, unmatched_simon, unmatched_chris
```

```python
results = []

for audio in shared_audios:
    df_audio = df[df['Audio File'] == audio]

    simon = df_audio[df_audio['Operator'] == 'Simon']
    chris = df_audio[df_audio['Operator'] == 'Chris']

    matches, unmatched_simon, unmatched_chris = match_annotations(simon, chris)

    total_matches = len(matches)
    exact_matches = sum(m['label_match'] for m in matches)

    results.append({
        'audio': audio,
        'simon_n': len(simon),
        'chris_n': len(chris),
        'matches': total_matches,
        'exact_label_matches': exact_matches,
        'label_accuracy': exact_matches / total_matches if total_matches > 0 else 0,
        'unmatched_simon': len(unmatched_simon),
        'unmatched_chris': len(unmatched_chris),
        'avg_start_diff': sum(m['start_diff'] for m in matches)/total_matches if total_matches else None,
        'avg_stop_diff': sum(m['stop_diff'] for m in matches)/total_matches if total_matches else None
    })

df_results = pd.DataFrame(results)
```
